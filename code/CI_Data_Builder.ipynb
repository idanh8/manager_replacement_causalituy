```python
import pandas as pd

def merge_teams_with_pl(teams, pl):
    # Convert dates to datetime format if they aren't already
    teams['fifa_update_date'] = pd.to_datetime(teams['fifa_update_date'])
    pl['Date'] = pd.to_datetime(pl['Date'])

    # Initialize an empty list to store the results
    merged_rows = []

    # Iterate through each row in the pl dataframe
    for _, pl_row in pl.iterrows():
        home_team = pl_row['HomeTeam']
        away_team = pl_row['AwayTeam']
        match_date = pl_row['Date']

        # Filter the teams dataframe for the home team and dates before the match date
        filtered_teams_home = teams[(teams['team_name'] == home_team) & (teams['fifa_update_date'] <= match_date)]
        filtered_teams_away = teams[(teams['team_name'] == away_team) & (teams['fifa_update_date'] <= match_date)]

        # Find the closest preceding dates for home and away teams
        if not filtered_teams_home.empty:
            closest_home_row = filtered_teams_home.loc[filtered_teams_home['fifa_update_date'].idxmax()]
        else:
            closest_home_row = None

        if not filtered_teams_away.empty:
            closest_away_row = filtered_teams_away.loc[filtered_teams_away['fifa_update_date'].idxmax()]
        else:
            closest_away_row = None

        # Prepare the merged row including all relevant data from both dataframes
        merged_row = pl_row.to_dict()  # Start with the pl row

        # Add the closest matching data for the home team
        if closest_home_row is not None:
            merged_row.update({f"home_{col}": closest_home_row[col] for col in teams.columns if col != 'team_name'})

        # Add the closest matching data for the away team
        if closest_away_row is not None:
            merged_row.update({f"away_{col}": closest_away_row[col] for col in teams.columns if col != 'team_name'})

        # Append the merged row to the results
        merged_rows.append(merged_row)

    # Convert the list of merged rows into a dataframe
    merged_df = pd.DataFrame(merged_rows)
    return merged_df

# Use the function to merge the dataframes
merged_df = merge_teams_with_pl(teams, pl)

# merged_df now contains all 1900 rows from pl with the relevant data from teams
```
